```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
StudentsPerformance = pd.read_csv("StudentsPerformance.csv")

# VIS 1: Barplot of race/ethnicity
x = StudentsPerformance['race/ethnicity'].sample(100, replace=True)
plt.bar(x.value_counts().index, x.value_counts(), color=["pink", "red", "blue", "green", "gold"])
plt.ylim(0, 30)
plt.title("Barplot of Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Count")
plt.show()

# VIS 2: Box plot for the scores
plt.boxplot([StudentsPerformance['math score'], StudentsPerformance['reading score'], StudentsPerformance['writing score']],
            labels=['Math', 'Reading', 'Writing'], patch_artist=True, boxprops=dict(facecolor='red', alpha=0.7))
plt.title("Box Plot for Test Scores")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.show()

# VIS 3: Grouped bar chart for test preparation course by race/ethnicity
df = pd.DataFrame({'Test Preparation Course': StudentsPerformance['test preparation course'],
                   'Race/Ethnicity': StudentsPerformance['race/ethnicity']})
sns.catplot(data=df, x='Test Preparation Course', hue='Race/Ethnicity', kind='count', palette='Set1')
plt.title("Grouped Bar Chart for Test Preparation Course by Race/Ethnicity")
plt.xlabel("Test Preparation Course")
plt.ylabel("Count")
plt.show()

# VIS 4: Histogram of Students' combined score
totalscore = StudentsPerformance['math score'] + StudentsPerformance['reading score'] + StudentsPerformance['writing score']
plt.hist(totalscore, bins=20, color='red', edgecolor='black')
plt.title("Histogram of Students' Combined Score")
plt.xlabel("Combined Score")
plt.ylabel("Frequency")
plt.show()

# VIS 5: Gender Pie Chart
gender_counts = StudentsPerformance['gender'].value_counts()
plt.pie(gender_counts, labels=['Female', 'Male'], colors=['magenta', 'red'], autopct='%1.1f%%')
plt.title("Gender Pie Chart")
plt.show()
```
