```python
import pandas as pd
import scipy.stats as stats

# Create a DataFrame with the given data
data = pd.DataFrame({
    'Height': [65.8, 71.5, 69.4, 68.2, 67.8, 68.7, 69.8, 70.1, 67.9, 66.8],
    'Weight': [112, 136, 153, 142, 144, 123, 141, 136, 112, 120],
    'Age': [30, 19, 45, 22, 29, 50, 51, 23, 17, 39],
    'Grip_Strength': [30, 31, 29, 28, 24, 26, 22, 20, 19, 31],
    'Frailty': ['N', 'N', 'N', 'Y', 'Y', 'N', 'Y', 'Y', 'N', 'N']
})

# Perform Welch's t-test
t_statistic, p_value = stats.ttest_ind(data[data['Frailty'] == 'N']['Age'], data[data['Frailty'] == 'Y']['Age'], equal_var=False)

# Calculate the confidence interval
confidence_interval = stats.t.interval(0.95, len(data[data['Frailty'] == 'N']) + len(data[data['Frailty'] == 'Y']) - 2,
                                       loc=data[data['Frailty'] == 'N']['Age'].mean() - data[data['Frailty'] == 'Y']['Age'].mean(),
                                       scale=stats.sem(data[data['Frailty'] == 'N']['Age'] + data[data['Frailty'] == 'Y']['Age']))

# Print the results
print(f"Welch Two Sample t-test:")
print(f"t = {t_statistic:.4f}, df = {len(data[data['Frailty'] == 'N']) + len(data[data['Frailty'] == 'Y']) - 2:.4f}, p-value = {p_value:.4f}")
print(f"95 percent confidence interval: {confidence_interval[0]:.5f} {confidence_interval[1]:.5f}")
print(f"sample estimates:")
print(f"mean in group N: {data[data['Frailty'] == 'N']['Age'].mean():.2f}")
print(f"mean in group Y: {data[data['Frailty'] == 'Y']['Age'].mean():.2f}")
```
